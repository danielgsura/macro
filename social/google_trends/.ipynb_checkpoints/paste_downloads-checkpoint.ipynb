```python
df = pd.DataFrame()
countries = []
for file in csv_files:
    country = file.split('.csv')[0].split('web_downloads\\')[1]
    countries.append(country)
    df_temp = pd.read_csv(file)
    df_temp = df_temp.iloc[1: , :]
    df_temp.columns = [country]
    print(country)
    df = pd.concat([df, df_temp], axis = 1)
    
```

```python
countries
```

```python
countries_dict = {'ar':'Google_Argentina', 'br':'Google_Brazil', 'cl':'Google_Chile', 'co':'Google_Colombia',
                  'mx':'Google_Mexico', 'pa':'Google_Panama', 'rd':'Google_Dominican Republic', 'sv':'Google_El Salvador',
                  'uy':'Google_Uruguay', 'fr':'Google_France', 'ht':'Google_Haiti',
                 'il':'Google_Israel','pe':'Google_Peru'}
```

```python
df = df.replace('<1', '1')
```

```python
for col in df.columns:
    df[col] = df[col].astype(int)
    df.rename(columns={col:countries_dict[col]}, inplace = True)
```

```python
df.reset_index(drop = False, inplace = True)
df.rename(columns = {'index':'date'}, inplace = True)
df
```
